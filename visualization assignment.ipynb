```python
# 1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.
import seaborn as sns
time = np.arange(100)
value=np.random.randn(100).cumsum()
df=pd.DataFrame({'time':time,'value':value})
sns.relplot(x = 'value',y='time',data = df)
```

```python
# 2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.
df = np.random.randn(10)
sns.histplot(df)
```

```python
# 3. Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.
time = np.arange(5)
value=np.random.randn(5).cumsum()
df=pd.DataFrame({'time':time,'value':value})
sns.catplot(x = 'value',y='time',data = df,kind = 'bar')
```

```python
# 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.
np.random.seed(42)
categories = ['Category A', 'Category B', 'Category C']
num_records_per_category = 100

data = {
    'Category': np.repeat(categories, num_records_per_category),
    'Value': np.concatenate([
        np.random.normal(loc=10, scale=5, size=num_records_per_category),
        np.random.normal(loc=20, scale=10, size=num_records_per_category),
        np.random.normal(loc=30, scale=15, size=num_records_per_category)
    ])
}


df = pd.DataFrame(data )
sns.boxplot(x='Category', y='Value', data=df)

```

```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

np.random.seed(42)
num_records = 100

feature_1 = np.random.normal(loc=50, scale=10, size=num_records)
feature_2 = feature_1 + np.random.normal(loc=0, scale=5, size=num_records)
feature_3 = feature_1 * 0.5 + np.random.normal(loc=0, scale=5, size=num_records)
feature_4 = feature_2 * 0.3 + np.random.normal(loc=0, scale=5, size=num_records)
data = {
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3,
    'Feature 4': feature_4
}
df = pd.DataFrame(data)
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt='.2f', linewidths=0.5, cbar=True)



```

```python
# 1. Using the given dataset,to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.
import plotly.graph_objects as go
import plotly.express as px

np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)
df
fig=go.Figure()
fig.add_trace(go.Scatter3d(x=df.X,y=df.Y,mode='markers',z=df.Z))
fig.show()
```

```python
# 2. Using the Student Grades, create a violin plot to display the distribution of scores across different grade categories.
np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)
df
fig = px.violin(df, y='Score', x='Grade')
fig.show()
```

```python
# 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
fig = px.imshow(df)
fig.show()
```

```python
# 4. Using the given x and y data, generate a 3D surface plot to visualize the function 
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
df
fig=go.Figure()
fig.add_trace(go.Surface(x=df.X,y=df.Y,))
fig.show()
```
